```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/michigan_housing.csv")

df["date"] = pd.to_datetime(df["date"])
df["month_num"] = range(len(df))

df.head()
```

```python
X = df[["month_num"]]
y = df["median_price"]

model = LinearRegression()
model.fit(X, y)

predictions = model.predict(X)

df["predicted_price"] = predictions
```

```python
plt.figure(figsize=(8,5))

plt.plot(df["date"], df["median_price"], label="Actual Price")
plt.plot(df["date"], df["predicted_price"], label="Predicted Price")

plt.xticks(rotation=45)
plt.title("Michigan Housing Price Forecast")
plt.legend()

plt.savefig("../outputs/figures/housing_forecast.png")
plt.show()
```
